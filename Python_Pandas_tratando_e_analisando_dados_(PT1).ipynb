## Tipos de Imóveis

```
dados.head(10)
```

```
#Selecionando uma coluna
tipo_de_imovel = dados['Tipo']
```

```
# visualizar o type da variável
type(tipo_de_imovel)
```

```
"""
- uma maneira de visualizar quais são os valores únicos de uma série. 
- diferentemente da "value_counts()", essa função não conta a ocorrência de cada
variável, mas sim apaga todos os valores duplicados, mantendo o index do lado 
esquerdo.
"""
tipo_de_imovel.drop_duplicates()
```

```
"""
- O primeiro parâmetro decide qual vai ser o valor da variável será mantido
caso seja first sera o primeiro.
- O segundo parametro decide se os valores duplicados serão apagados
"""
tipo_de_imovel.drop_duplicates(keep='first', inplace=True)
tipo_de_imovel
```

## Ajustando a visualização

```
"""
Apesar de ter sido transformado em um dataframe, o index não está bom para uma 
apresentação, pois corresponde a cada primeira ocorrência de cada variável
"""
tipo_de_imovel = pd.DataFrame(tipo_de_imovel)
tipo_de_imovel
```

```
# para isso precisamos saber qual o tamanho do index, e assim substituir
tipo_de_imovel.index = range(len(tipo_de_imovel.index))
```

```
tipo_de_imovel
```

## (EXTRA II) criando e concatenando dataframes

```
# primeiramente criei uma lista com 3 listas dentro
lista = [[1, 2, 3], 
         [4, 5, 6], 
         [7, 8, 9]]
lista
```

```
# a tranformo em um dataframe
df1 = pd.DataFrame(data = lista)
```

```
df1
```

```
# devo arrumar os índices, então primeiramente cou criar uma variavel para ele
index = ["linha_" + str(i) for i in range (3)]
index
```

```
df1.index = index
df1
```

```
#agora farei o mesmo para as colunas
columns = ["coluna_" + str(i) for i in range(3)]
columns
```

```
df1.columns = columns
df1
```

```
# É possível fazer o mesmo com um dicionário
dicionario = {
    'coluna_0': {'linha_0': 1, 'linha_1': 4, 'linha_2': 7},
    'coluna_1': {'linha_0': 2, 'linha_1': 5, 'linha_2': 8},
    'coluna_2': {'linha_0': 3, 'linha_1': 6, 'linha_2': 9}
}

df2 = pd.DataFrame(data = dicionario)
df2
```

```
# posso fazer também com uma lista de tuplas
lista_tuplas = [(1, 2, 3), 
                (4, 5, 6), 
                (7, 8, 9)]
lista_tuplas
```

```
"""
agora criei pelo metodo .dataframe, com as variaveis já criadas anteriormente
mas agora dentro dos parâmetros
"""
df3 = pd.DataFrame(data=lista_tuplas, columns=columns, index=index)
df3
```

```
"""
- agora vou concatenar os dataframes
- para isso, irei mudar os valores dentro dos dataframes, para a concatenação 
ficar mais clara
"""
df1[df1 > 0] = 'A'
df1
```

```
df2[df2 > 0] = 'B'
df2
```

```
df3[df3 > 0] = 'C'
df3
```

```
df4 = pd.concat([df1, df2, df3])
df4
```

```
"""
- nota-se no exemplo anterior que os valores anteriores foram 
ligados pelas colunas
- é possível também fazer que os valores concatenados sejam ligador pelas linhas
"""

df5 = pd.concat([df1, df2, df3], axis = 1)
df5
```
